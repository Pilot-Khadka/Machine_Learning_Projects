# Principal Component Analysis(PCA) for facial recognition

### First, to get familiar, performing PCA for 2-D data

```python
import matplotlib.pyplot as plt
import random

# Generate random points in the first quadrant with small variance in the y direction
num_points = 20
x = [random.uniform(0, 10) for _ in range(num_points)]
y = [random.uniform(0, 10) for _ in range(num_points)]

# Create the plot
plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('2D Points in the First Quadrant with Small Variance in Y Direction')

# Set the x and y axis limits
plt.xlim(0, 10)
plt.ylim(0, 10)

# Display the plot
plt.show()
```

```python
import random
import numpy as np
import matplotlib.pyplot as plt

X = np.abs(np.random.randn(20, 2))

plt.axhline(0, color = 'black') 
plt.axvline(0, color = 'black')
plt.title('20 random points')
plt.xlim(-3,3)
plt.ylim(-3,3)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.scatter([X[:,0]],[X[:,1]])
plt.text(0.9, 0.1, 'Roll: 18, 25', ha='right', va='bottom', transform=plt.gca().transAxes,fontsize=14)
plt.grid()
```

```python
'''
Subtraction of mean from the raw data to shift the center of the data to the origin
This does not change the relative position of data points to each other
'''

X = X-np.mean(X,axis=0)

plt.axhline(0, color = 'black') 
plt.axvline(0, color = 'black') 
plt.title('Centered data')
plt.xlim(-3,3)
plt.ylim(-3,3)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.scatter([X[:,0]],[X[:,1]])
plt.grid()
plt.text(0.9, 0.1, 'Roll: 18, 25', ha='right', va='bottom', transform=plt.gca().transAxes,fontsize=14)
```

```python
mat = np.random.rand(2, 2)
y = np.dot(X, mat)

plt.axhline(0, color = 'black') 
plt.axvline(0, color = 'black') 
plt.title('Centered data')
plt.xlim(-2,2)
plt.ylim(-2,2)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.scatter([y[:,0]],[y[:,1]])
plt.grid()
plt.text(0.9, 0.1, 'Roll: 18, 25', ha='right', va='bottom', transform=plt.gca().transAxes,fontsize=14)
```

```python
cov = np.cov(np.transpose(y))
eigen_values, eigen_vectors = np.linalg.eig(cov)

plt.title("Data with eigen vectors")
plt.axhline(0, color = 'black') 
plt.axvline(0, color = 'black') 
plt.xlim(-2,2)
plt.ylim(-2,2)
plt.scatter(y[:, 0], y[:, 1], alpha=0.5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.text(0.9, 0.1, 'Roll: 18, 25', ha='right', va='bottom', transform=plt.gca().transAxes,fontsize=14)
plt.grid()

# Plot the eigenvectors
for eigenvector in eigen_vectors.T:
    plt.arrow(0, 0, eigenvector[0], eigenvector[1], head_width=0.1, head_length=0.1, fc='red', ec='red')
```
