## **Principal Component Analysis on Fashion-MNIST dataset**

```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt

# Load Fashion MNIST dataset
fashion_mnist = tf.keras.datasets.fashion_mnist
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()

# Print shapes of the loaded data
print("Train images shape:", train_images.shape)
print("Train labels shape:", train_labels.shape)
print("Test images shape:", test_images.shape)
print("Test labels shape:", test_labels.shape)

# Visualize a sample image from the training dataset
plt.figure()
plt.imshow(train_images[0])
plt.colorbar()
plt.grid(False)
plt.show()
```

```python
# Find and print unique labels
unique_labels = np.unique(train_labels)
print("Unique Labels:")
print(unique_labels)
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

label = 2  # Example label
print("Class name for label", label, ":", class_names[label])
```

```python
# Iterate through each class and find an example image and its label
for i in range(len(class_names)):
    index = np.where(train_labels == i)[0][0]
    
    # Get the image and corresponding label name
    example_image = train_images[index]
    label_name = class_names[i]
    
    plt.figure()
    plt.imshow(example_image, cmap=plt.cm.binary)
    plt.xlabel(label_name)
    plt.colorbar()
    plt.grid(False)
    plt.show()
```

```python
# Normalize the pixel values to be between 0 and 1
train_images_normalized = train_images / 255.0
test_images_normalized = test_images / 255.0

# Reshape the images to a single row (1D array)
train_images_flat = train_images.reshape(train_images.shape[0], -1)
test_images_flat = test_images.reshape(test_images.shape[0], -1)

# Print the shapes after reshaping to a single row
print("Train images shape after reshaping to a single row:", train_images_flat.shape)
print("Test images shape after reshaping to a single row:", test_images_flat.shape)
```

## Normalize the Data by Subtracting the Mean Values of Each Dimension

```python
# Calculate the mean values of each feature (pixel column)
mean_values = np.mean(train_images_flat, axis=0)

# Normalize the data by subtracting the mean values from the flattened images
normalized_train_images = train_images_flat - mean_values

# Example: Access the normalized data of the 5th image (index 4)
example_normalized_image = normalized_train_images[4]
example_normalized_image_reshaped = example_normalized_image.reshape(28, 28)

# Plot the normalized image of the 5th sample
plt.figure()
plt.imshow(example_normalized_image_reshaped, cmap='gray')
plt.colorbar()
plt.grid(False)
plt.show()
```

## Calculate covariance matrix of data matrix

```python
# Size of data matrix: (60000, 784)
# Covariance matrix size must be 784x784

# Calculate the covariance matrix
cov_matrix = np.cov(normalized_train_images, rowvar=False)

print("Covariance matrix shape:", cov_matrix.shape)

# Visualizing the covariance matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(cov_matrix, cmap='coolwarm', annot=False, fmt=".2f", cbar=True, square=True)
plt.title('Covariance Matrix')
plt.show()
```

## Calculate eigenvectors and eigenvalues of covariance matrix

```python
# Calculate the eigenvectors and eigenvalues
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Sort the eigenvalues and corresponding eigenvectors in descending order
sorted_indices = np.argsort(eigenvalues)[::-1]  
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]
```

## Select top k eigenvectors as basis vectors

```python
# Select the top 10 eigenvectors
top_10_eigenvectors = eigenvectors[:, :10]
top_10_transposed = np.transpose(top_10_eigenvectors)

# Create a subplot grid for displaying the eigen images
fig, axs = plt.subplots(2, 5, figsize=(12, 6))
fig.suptitle('Top 10 Eigen Images')

# Iterate over the top 10 eigen vectors
for i, ax in enumerate(axs.flat):
    eigen_image = np.real(top_10_transposed[i].reshape((28, 28)))
    ax.imshow(eigen_image, cmap='gray')
    ax.set_title(f"Eigen #{i+1}")
    ax.axis('off')

plt.tight_layout()
plt.show()
```

## Calculate proportion of variance

```python
# Calculate the eigenvectors and eigenvalues
eigenvalues, _ = np.linalg.eig(cov_matrix)

# Sort eigenvalues in descending order
sorted_eigenvalues = np.sort(eigenvalues)[::-1]

# Calculate the total sum of eigenvalues
total_variance = np.sum(sorted_eigenvalues)

# Calculate the proportion of variance explained by each eigenvalue
proportion_of_variance = sorted_eigenvalues / total_variance

# Plot the proportion of variance vs eigenvalues
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(proportion_of_variance) + 1), np.cumsum(proportion_of_variance), marker='o', linestyle='-', color='b')
plt.xlabel('Number of Principal Components')
plt.ylabel('Proportion of Variance Explained')
plt.title('Proportion of Variance vs Eigenvalues')
plt.grid(True)
plt.show()
```

## New dataset using principal components


```python
# Number of top principal components to use
k = 10

# Calculate the top k eigenvectors
top_k_eigenvectors = eigenvectors[:, :k]

# Project the original data onto the top k eigenvectors to get the new data
new_data = np.dot(normalized_train_images, top_k_eigenvectors)

# Calculate the covariance matrix of the new data
new_cov_matrix = np.cov(new_data, rowvar=False)

# Plot the covariance matrix of the new data
plt.figure(figsize=(8, 8))
sns.heatmap(new_cov_matrix, cmap='coolwarm', annot=False, fmt=".2f", cbar=True, square=True)
plt.title('Covariance Matrix of New Data (Top 10 Principal Components)')
plt.show()
```

```python
# Create a subplot grid for displaying covariance matrices in a single figure
fig, axs = plt.subplots(3, 3, figsize=(15, 15))
fig.suptitle('Covariance Matrices of New Data (Top 2 to 10 Principal Components)')

# Iterate over values of k from 2 to 10
for k, ax in zip(range(2, 11), axs.flat):
    top_k_eigenvectors = eigenvectors[:, :k]
    new_data = np.dot(normalized_train_images, top_k_eigenvectors)
    new_cov_matrix = np.cov(new_data, rowvar=False)
    
    sns.heatmap(new_cov_matrix,cmap='coolwarm', annot=False, fmt=".2f", cbar=True,square=True, ax=ax)
    ax.set_title(f'Top {k} Principal Components')
    ax.axis('off')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

## Reconstructing Original data from the transformed data

```python
# Select a single image for reconstruction (e.g., the first image)
image_to_reconstruct = normalized_train_images[0]

fig, axs = plt.subplots(1, 9, figsize=(15, 5))
fig.suptitle('Reconstruction of a Single Image from Top 2 to Top 10 Principal Components')

# Iterate over values of k from 2 to 10
for k, ax in zip(range(2, 11), axs):
    top_k_eigenvectors = eigenvectors[:, :k]
    transformed_data = np.dot(image_to_reconstruct, top_k_eigenvectors)
    reconstructed_data = np.dot(transformed_data, top_k_eigenvectors.T)
    final_reconstructed_image = reconstructed_data + mean_values

    ax.imshow(final_reconstructed_image.reshape(28, 28), cmap='gray')
    ax.set_title(f'Top {k}')
    ax.axis('off')

plt.tight_layout(rect=[0, 0.03, 1, 0.9])
plt.show()
```

## Calculate the reconstruction errors for different numbers of principal components


```python
reconstruction_errors = []

for k in range(2, 31):
    top_k_eigenvectors = eigenvectors[:, :k]
    transformed_data = np.dot(normalized_train_images, top_k_eigenvectors)
    reconstructed_data = np.dot(transformed_data, top_k_eigenvectors.T)
    final_reconstructed_images = reconstructed_data + mean_values

    # Calculate mean squared error as reconstruction error
    mse = np.mean((normalized_train_images - final_reconstructed_images) ** 2)
    reconstruction_errors.append(mse)

plt.figure(figsize=(8, 6))
plt.plot(range(2, 31), reconstruction_errors, marker='o', linestyle='-', color='b')
plt.xlabel('Number of Principal Components')
plt.ylabel('Reconstruction Error (MSE)')
plt.title('Reconstruction Error vs Number of Principal Components')
plt.grid(True)
plt.show()
```

## Comparision of Model trained in full dataset vs top-50 principal components in pytorch

```python
import torch
from torch.utils.data import DataLoader
from torchvision import datasets, transforms

# Define a transform to normalize the data
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,), (0.5,))])

# Load the full dataset
full_dataset = datasets.FashionMNIST(root='./data', train=True, download=True, transform=transform)

# Load the top 50 principal components (already calculated and stored in top_50_principal_components)
top_50_principal_components = eigenvectors[:, :50]
transformed_data = torch.tensor(np.dot(normalized_train_images, top_50_principal_components), dtype=torch.float32)

# Combine transformed data with labels to create a custom dataset
class CustomDataset(torch.utils.data.Dataset):
    def __init__(self, transformed_data, targets):
        self.data = transformed_data
        self.targets = targets

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        return self.data[idx], self.targets[idx]

# Create datasets and data loaders for both cases
full_dataset_loader = DataLoader(full_dataset, batch_size=64, shuffle=True)
transformed_dataset = CustomDataset(transformed_data, torch.tensor(train_labels))
transformed_loader = DataLoader(transformed_dataset, batch_size=64, shuffle=True)
```

```python
import torch.nn as nn
import torch.optim as optim

# Define a simple neural network model
class SimpleNN(nn.Module):
    def __init__(self, input_size, output_size):
        super(SimpleNN, self).__init__()
        self.fc = nn.Linear(input_size, output_size)

    def forward(self, x):
        x = self.fc(x)
        return x

# Initialize the models
full_dataset_model = SimpleNN(input_size=784, output_size=10)
transformed_model = SimpleNN(input_size=50, output_size=10)

# Define loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer_full_dataset = optim.Adam(full_dataset_model.parameters(), lr=0.001)
optimizer_transformed = optim.Adam(transformed_model.parameters(), lr=0.001)

# Training loop for the full dataset model
for epoch in range(5):  # train for 5 epochs as an example
    for inputs, labels in full_dataset_loader:
        optimizer_full_dataset.zero_grad()
        outputs = full_dataset_model(inputs.view(inputs.size(0), -1))
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer_full_dataset.step()

# Training loop for the transformed dataset model
for epoch in range(5):  # train for 5 epochs as an example
    for inputs, labels in transformed_loader:
        optimizer_transformed.zero_grad()
        outputs = transformed_model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer_transformed.step()
```

```python
from sklearn.metrics import classification_report
import torch

# Set the models to evaluation mode
full_dataset_model.eval()
transformed_model.eval()

# Define a function to get predictions from the model
def get_predictions(model, dataloader):
    all_preds = []
    all_labels = []
    with torch.no_grad():
        for inputs, labels in dataloader:
            outputs = model(inputs.view(inputs.size(0), -1))
            _, preds = torch.max(outputs, 1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
    return all_preds, all_labels

# Get predictions from both models
full_dataset_preds, full_dataset_labels = get_predictions(full_dataset_model, full_dataset_loader)
transformed_preds, transformed_labels = get_predictions(transformed_model, transformed_loader)
```

```python
# Import necessary libraries
from sklearn.metrics import classification_report

# Print classification report for the full dataset model
print("Classification Report for Full Dataset Model:")
print(classification_report(full_dataset_labels, full_dataset_preds))

# Print classification report for the transformed dataset model
print("Classification Report for Transformed Dataset Model:")
print(classification_report(transformed_labels, transformed_preds))
```
