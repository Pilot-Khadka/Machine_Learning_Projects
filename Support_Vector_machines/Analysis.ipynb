```python
from SVM import SVM
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification

x,y = make_classification(n_samples=500, n_features=2,n_informative=2, n_redundant=0, n_clusters_per_class=1)
print(len(x))
print(len(y))
plt.scatter(x[:,0], x[:,1], c=y)
```

SVM from implemenation

```python
X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.33, random_state=100
)
model = SVM(kernel_type='linear')
model.fit(X_train, y_train)
pred = model.predict(X_test)
accuracy = model.calculate_accuracy(pred,y_test)
print(accuracy)
```

SVM from sklearn

```python
from sklearn import svm
from sklearn.metrics import accuracy_score

sklearn_model = svm.SVC()
sklearn_model.fit(X_train,y_train)
y_pred = sklearn_model.predict(X_test)
sklearn_accuracy = accuracy_score(y_test,y_pred)
print(sklearn_accuracy)
```
