```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 200)  
y = np.linspace(-10, 10, 200)  

X, Y = np.meshgrid(x, y)  # Create a grid of coordinates

# Define a range of p values
p_values = [0.0025, 0.6, 1, 2, 3, 25, 123, 568, 3000]

num_rows = 3
num_cols = 3

fig, axs = plt.subplots(num_rows, num_cols, figsize=(12, 10))


for i,p in enumerate(p_values):
    
    abs_X_pow_p = np.abs(X) ** p
    abs_Y_pow_p = np.abs(Y) ** p
    Z = (abs_X_pow_p + abs_Y_pow_p) ** (1 / p)
    
    row_index = i // num_cols
    col_index = i % num_cols
    
    axs[row_index, col_index].contour(X, Y, Z, levels=[1])
    axs[row_index, col_index].set_title(f'p = {p:.4f}')
    
    axs[row_index, col_index].axhline(0, color='black')
    axs[row_index, col_index].axvline(0, color='black')
    axs[row_index, col_index].set_xlim(-2, 2)
    axs[row_index, col_index].set_ylim(-2, 2)
    
plt.show()
```
